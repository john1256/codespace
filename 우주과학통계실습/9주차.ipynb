```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n_coins = 10
n_trial = 1000
x = np.random.randint(0,2,size = (n_coins, n_trial))
a = np.sum(x, axis = 0)
mean = a.mean()
std = a.std()
```

```python
r = stats.norm(loc = mean, scale = std)
x = np.linspace(0,10,100)
y = r.pdf(x)
```

```python
hist, bin_edges = np.histogram(a, bins = np.arange(0,12,1))
plt.hist(a, bins= bin_edges, density = True, align = 'left')
plt.plot(x,y)
```

```python
from scipy.special import factorial
```

```python
def binomial(n,p,k):
    y = factorial(n)/factorial(n-k)/factorial(k)*np.power(p,k)*np.power(1-p,n-k)
    return y
p = 0.25
n_trials = [3,10,20,50]
for n_trial in n_trials:
    x = np.arange(0,n_trial+1,1)
    y = binomial(n_trial,p,x)
    
    avg = n_trial*p
    std = np.sqrt(n_trial*p*(1-p))
    
    x2 = np.linspace(0,n_trial,100)
    y2 = stats.norm(loc = avg, scale = std).pdf(x2)
    plt.bar(x,y)
    plt.plot(x2,y2,'--', c = 'red')
    plt.show()
```

```python
def poisson(x,u):
    result = np.power(u,x)/factorial(x)*np.exp(-u)
    return result

x = np.arange(0,11,1)
y = poisson(x,3)
    
```
