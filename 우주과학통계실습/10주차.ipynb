```python
import numpy as np
from scipy.special import factorial
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
np.power(10,20)

```

```python
def poisson(x,u):
    result = np.power(u,x)/factorial(x)*np.exp(-u)
    return result


u = [1,3, 5,10,20]
x = np.arange(0,30,1, dtype=float)
xspace = np.linspace(0,30,100)
for i in u:
    y = poisson(x,i)
    plt.bar(x,y, label = 'Poisson')
    x_rv = stats.norm(loc = i, scale = np.sqrt(i))
    
    ynorm = x_rv.pdf(xspace)
    plt.plot(xspace,ynorm, '--',color = 'r',label = 'Gaussian')
    plt.legend()
    plt.title(r'Comparison of Poisson & normal distribution of $\mu$ = 3')
    plt.show()
    
```

```python
#1

from scipy.stats import poisson
#sample mean of Poisson dist
mu = 3
rv = poisson(mu)
n = 20000
sample_size = 5000
x_sample = rv.rvs((n,sample_size))
sample_mean = np.mean(x_sample,axis=1)

#normal dist with avg = mu, sigma = np.sqrt(mu/sample_size)

x_sig = np.sqrt(mu/sample_size)
rv_norm = stats.norm(mu,x_sig)
x1 = np.linspace(mu-3*x_sig, mu+3*x_sig,100)
y1 = stats.norm.pdf(x1, loc =mu, scale = x_sig)

plt.hist(sample_mean, bins = 100, label = 'poisson', density=True)
plt.plot(x1,y1, '--', color = 'r', label = 'normal')
plt.legend()
```

```python
#2 

#uniform random number
n = 10000
sample_size = 1000
rv = stats.uniform(loc = 0,scale = 1)
x_sample = rv.rvs((n,sample_size))
sample_mean = np.mean(x_sample, axis = 1)
mean = np.mean(sample_mean)
sample_std = np.std(sample_mean)

x1 = np.linspace(mean - 3*sample_std, mean + 3*sample_std, 100)
y1 = stats.norm.pdf(x1, loc = mean, scale = sample_std)

plt.hist(sample_mean, bins = 100, label = 'random', density=True)
plt.plot(x1, y1, '--', color = 'red', label = 'gaussian')
plt.legend()
```
